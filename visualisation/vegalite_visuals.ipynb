```python
from vega import VegaLite
import pandas as pd
```

```python
colnames = ['Entities','Book','Count']
df = pd.read_csv('data\melt_old_per.csv', names=colnames, index_col=0,skiprows=1)
df.head()
```

```python
VegaLite({
    "title": "Top three extracted PER entities across the Old Testament books",
    "width": 800,
  "height": 400,
  "mark": "area",
  "encoding": {
    "x": {
      "field": "Book", "type": "quantitative",
      "axis": {"domain": "true", "tickSize": 10, "fontSize": 30}
    },
    "y": {
      "aggregate": "sum", "field": "Count","type": "quantitative",
      "axis": {"domain": "true", "tickSize": 10, "fontSize": 20},
      "stack": "center"
    },
    "color": {"field":"Entities", "type":"nominal", "fontSize": 20,"scale":{"scheme": "category20c"},"legend": {"columns": 1}
}
  }
}, df)
```
